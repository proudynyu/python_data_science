```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

x_train = np.sort(np.random.rand(50) * 100)
y_train = 0.5 * x_train + np.random.normal(0, 5, 50)

def manual_knn_predict(x_query, x_data, y_data, k):
    distances = np.abs(x_data - x_query)
    nearest_idx = np.argsort(distances)[:k]
    prediction = np.mean(y_data[nearest_idx])
    return prediction
```

```python
x_range = np.linspace(0, 100, 500)
k_value = 3

y_preds = [manual_knn_predict(xq, x_train, y_train, k_value) for xq in x_range]

plt.figure(figsize=(10,6))
plt.scatter(x_train, y_train, color="black", label="Original Data (Neighbors)")
plt.plot(x_range, y_preds, color="red", linewidth=2, label=f"KNN Prediction (k={k_value})")
```
